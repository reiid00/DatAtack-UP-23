```python
anpc[['Natureza1', 'Natureza2', 'Natureza3']] = anpc['Natureza'].str.split(' / ', expand=True).drop(3,axis=1)
```

```python
# create a dictionary to map the original labels to the new groupings
mapping = {
    'Assistência em Saúde': 'Assistência e Prevenção',
    'Assistência e Prevenção a actividades humanas': 'Assistência e Prevenção',
    'Incêndios Rurais': 'Incêndios',
    'Acidentes': 'Acidentes',
    'Intervenção em conflitos legais': 'Segurança e Serviços',
    'Comprometimento total ou parcial de segurança, serviços ou estruturas': 'Segurança e Serviços',
    'Incêndios Urbanos ou em Área Urbanizável': 'Incêndios',
    'Incêndios em Detritos': 'Incêndios em Bens',
    'Incêndios em Transportes': 'Incêndios em Bens',
    'Incêndios em Equipamento e Produtos': 'Incêndios em Bens',
    'Operações': 'Operações',
    'Acidentes industriais e tecnológicos': 'Segurança e Serviços',
    'Fenómenos Naturais': 'Fenómenos Naturais'
}

# apply the mapping to create a new column with the grouped labels
anpc['Grouped'] = anpc['Natureza2'].map(mapping)
```

```python
anpc = anpc.dropna(subset=['Grouped'])
anpc['Grouped'].unique()
```

```python
anpc['Grouped'].value_counts()
```

```python
anpc.groupby('Grouped')['NumeroOperacionaisTerrestresEnvolvidos'].sum()
```

```python
# remove the rows with the value 'B' in the "Column B" column
anpc = anpc[anpc['Grouped'] != 'Fenómenos Naturais']
```

```python
anpc.groupby('Grouped')['NumeroOperacionaisTerrestresEnvolvidos'].sum()
```

```python
anpc=anpc.drop('Natureza1', axis=1)
```
